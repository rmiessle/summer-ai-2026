## Working With Structured Data 
This portion of the session focuses on a [set of letters](https://jackpeirs.org) written from 1915-1919 by H.J.C. (Jack) Peirs, a British World War I officer.

The data we are working with are in spreadsheet, with rows that correspond to each letter, and columns that identify the text of the letter (transcribed by a human, not OCR), the date it was sent, and who received it. This is a type of **structured data**.

## Loading the Letters Dataset

The file `letters.csv` is in the same folder as this notebook. When we load it, we will see the first few rows. We should also see the column names, `text`, `recipient`, and ``date``.

This code also loads four R libraries (also called packages), which are packages of code that extend the basic functionality of R. These packages focus on analyzing text from a data science approach.

Run the next cell to load it by clicking in the cell and pressing **Shift + Enter**.



```r
# loads R libraries for the activity
#----
# Loads the tidyverse collection of packages, including dplyr, ggplot2, readr, tidyr, and others.
# Useful for importing, cleaning, transforming, and visualizing data.
library(tidyverse)
#----
# Provides tools for text mining using tidy data principles.
# Useful for breaking text into words, counting terms, and preparing text for analysis.
library(tidytext)
#----
# Helps work with dates and times.
# Useful for parsing dates, extracting years/months/days, and organizing data chronologically.
library(lubridate)
#----
# Provides sentiment analysis tools.
# Useful for scoring text based on emotional tone, sentiment, and narrative shape.
library(syuzhet)
#----
# Text mining package with tools for cleaning and preparing text.
# Useful for removing punctuation, stop words, numbers, and extra whitespace.
library(tm)
#----
# Provides word stemming tools.
# Useful for reducing words to their root form, such as 'running,' 'runs,' and 'ran' to 'run.'
library(SnowballC)
#----
# Provides color palettes.
# Useful for creating visually consistent charts, graphs, and word clouds.
library(RColorBrewer)
#----
# Creates word clouds from text data.
# Useful for visually showing the most frequent words in a body of text.
library(wordcloud)
#----
# Provides tools for splitting text into tokens, such as words, sentences,
# paragraphs, lines, n-grams, and other text units.
# Useful when you need more control over how text is broken apart before analysis.
library(tokenizers)
#----
# load letters.csv and create a variable `letters`
letters <- read_csv("letters.csv", show_col_types = FALSE)
# review the dataset
glimpse(letters)
head(letters)
```
